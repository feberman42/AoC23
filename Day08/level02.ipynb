Open the Input and read lines as a list.  
Extract Instructions and pop empty lines.

```python
input = open('input')
lines = input.readlines()
instruction = lines[0].strip('\n')
lines.pop(0)
lines.pop(0)
```

Parse the input into a list of nodes[node[left, right]].

```python

nodes = [i.strip('\n').split(' = ') for i in lines]
nodes[0:5]
for node in nodes:
	node[1] = node[1].strip('()').split(', ')
nodes
```

Define helper to get idex of node by string.

```python
def	find_node(_node: str, nodes: list[str, list[str]]) -> int:
	for node in nodes:
		if (node[0] == _node):
			return (nodes.index(node))
```

```python
_node = find_node('AAA', nodes)
i = 0
while (nodes[_node][0] != 'ZZZ'):
	for d in instruction:
		if (d == 'L'):
			_node = find_node(nodes[_node][1][0], nodes)
		elif (d == 'R'):
			_node = find_node(nodes[_node][1][1], nodes)
		i += 1
		if (nodes[_node][0] == 'ZZZ'):
			break
i
```
