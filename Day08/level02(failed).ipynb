Open the Input and read lines as a list.  
Extract Instructions and pop empty lines.

```python
input = open('input')
lines = input.readlines()
instruction = lines[0].strip('\n')
lines.pop(0)
lines.pop(0)
```

```python
instructions = [int(i) for i in instruction.replace('L', '0').replace('R', '1')]
```

Parse the input into a list of nodes[node[left, right]].

```python

nodes = [i.strip('\n').split(' = ') for i in lines]
nodes[0:5]
for node in nodes:
	node[1] = node[1].strip('()').split(', ')
nodes
```

```python
def	find_node(_node: str, nodes: list[str, list[str]]) -> int:
	for i in range(len(nodes)):
		if (nodes[i][0] == _node):
			return (i)
```

```python
nodes_opt = []
for node in nodes:
	nodes_opt.append([find_node(node[1][0], nodes), find_node(node[1][1], nodes)])
nodes_opt
```

Define helper to get idex of node by string.

```python
start = []
for node in nodes:
	if (node[0][-1] == 'A'):
		start.append(nodes.index(node))
start = start[0:2]
start
```

```python
targets = []
for node in nodes:
	if (node[0][-1] == 'Z'):
		targets.append(nodes.index(node))
targets
```

```python
def	check_nodes(_nodes: list[int], targets: list[int]) -> bool:
	for _node in _nodes:
		if (_node not in targets):
			return False
	return True
```

```python
i = 0
_nodes = start
while (check_nodes(_nodes, targets) == False):
	for d in instructions:
		_nodes = [nodes_opt[k][d] for k in _nodes]
		i += 1
		if (check_nodes(_nodes, targets)):
			break
print(i, _nodes)
```
